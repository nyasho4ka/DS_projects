### Step 5. Testing hypotheses (Python)

```python
#rides = pd.read_csv('project_sql_result_07.csv')

rides = pd.read_csv('/datasets/project_sql_result_07.csv')
rides.head()
```

```python
rides.info()
```

```python
#Let's convert start_ts column to datetime
rides['start_ts'] = pd.to_datetime(rides['start_ts'], format='%Y-%m-%d %H:%M:%S')
```

```python
rides['start_ts'].dtype
```

```python
rides.head()
```

```python
# Now let's form query to the table and get all rides on rainy Saturdays
sunny_rides = rides.query('weather_conditions == "Good"')
rainy_rides = rides.query('weather_conditions == "Bad"')
```

```python
def test_means_of_2_populations(sample_1, sample_2, alpha=0.05):
    results = st.ttest_ind(sample_1.dropna(), sample_2.dropna())
    
    print('p-value: {}'.format(results.pvalue))
    
    if results.pvalue < alpha:
        print('We reject the null hypothesis')
    else:
        print("We can't reject the null hypothesis")
```

Let's form the null hypothesis:

$H_{0}$: The average duration of rides from the Loop to O'Hare International Airport is the same for rainy and sunny Saturdays

$H_{1}$: The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays

```python
# Let's check duration distribution for sunny and rainy Saturdays
sunny_durations = sunny_rides['duration_seconds']
rainy_durations = rainy_rides['duration_seconds']

sunny_durations.hist()
```

```python
rainy_durations.hist()
```

```python
test_means_of_2_populations(
    sunny_rides['duration_seconds'],
    rainy_rides['duration_seconds'])
```

### Conslusion

It seems that the means of trip duration on sunny and rainy days differ a lot. So we can reject the null hypothesis and say that in trips are longer in rainy Saturdays.
